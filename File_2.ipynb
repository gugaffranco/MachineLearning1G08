### Handle missing values

- Mean → for normally distributed data

- Median → for skewed data (more robust to outliers)

```python
#Change the missing values for median (numerical), normal distribution

learn_data_copy['ambient_humidity'] = learn_data_copy['ambient_humidity'].fillna(learn_data_copy['ambient_humidity'].mean())

learn_data_copy['cooling_period'] = learn_data_copy['cooling_period'].fillna(learn_data_copy['cooling_period'].mean())

learn_data_copy['cream_fat_content'] = learn_data_copy['cream_fat_content'].fillna(learn_data_copy['cream_fat_content'].mean())

learn_data_copy['egg_yolk_count'] = learn_data_copy['egg_yolk_count'].fillna(learn_data_copy['egg_yolk_count'].mean())

learn_data_copy['final_temperature'] = learn_data_copy['final_temperature'].fillna(learn_data_copy['final_temperature'].mean())

learn_data_copy['lemon_zest_ph'] = learn_data_copy['lemon_zest_ph'].fillna(learn_data_copy['lemon_zest_ph'].mean())

learn_data_copy['oven_temperature'] = learn_data_copy['oven_temperature'].fillna(learn_data_copy['oven_temperature'].mean())

learn_data_copy['preheating_time'] = learn_data_copy['preheating_time'].fillna(learn_data_copy['preheating_time'].mean())
```

```python
#Change the missing values for median (numerical), skewed data

learn_data_copy['egg_temperature'] = learn_data_copy['egg_temperature'].fillna(learn_data_copy['egg_temperature'].median())

learn_data_copy['baking_duration'] = learn_data_copy['baking_duration'].fillna(learn_data_copy['baking_duration'].median())

learn_data_copy['salt_ratio'] = learn_data_copy['salt_ratio'].fillna(learn_data_copy['salt_ratio'].median())

learn_data_copy['sugar_content'] = learn_data_copy['sugar_content'].fillna(learn_data_copy['sugar_content'].median())

learn_data_copy['vanilla_extract'] = learn_data_copy['vanilla_extract'].fillna(learn_data_copy['vanilla_extract'].median())
```

```python
# Mudar todas as "Lisboa", "Porto" e "Pastel de Nata" diferentes para iguais
```

```python
learn_data_copy['origin'] = learn_data_copy['origin'].str.strip().str.lower() # Take out the spaces and in lower case

learn_data_copy['origin'] = learn_data_copy['origin'].str.strip().str.title() # Capitalize


learn_data_copy['pastry_type'] = learn_data_copy['pastry_type'].str.strip().str.lower() # Take out the spaces and in lower case

learn_data_copy['pastry_type'] = learn_data_copy['pastry_type'].str.strip().str.title() # Capitalize

#Falta o pastel de nata

learn_data_copy['pastry_type'] = (learn_data_copy['pastry_type'].str.lower().str.strip())
learn_data_copy['pastry_type'] = learn_data_copy['pastry_type'].replace({'Pastel nata': 'pastel de nata','Pastel Nata': 'pastel de nata'})
learn_data_copy['pastry_type'] = learn_data_copy['pastry_type'].replace({'pastel nata': 'Pastel de Nata',})



# Verify
print(learn_data_copy['origin'].value_counts())

learn_data_copy['pastry_type'].value_counts()
```

```python
#Change the missing values for the mode (categorical)

#Isto tem de ser depois de tratar do Pastel de Nata

learn_data_copy['origin'] = learn_data_copy['origin'].fillna(learn_data_copy['origin'].mode()[0])

learn_data_copy['pastry_type'] = learn_data_copy['pastry_type'].fillna(learn_data_copy['pastry_type'].mode()[0])
```

```python
#Delete the whole column if it has a lot of missing values

if "notes_baker" in learn_data_copy.columns:
    learn_data_copy.drop(["notes_baker"], axis=1, inplace=True)
```

```python
# Delete the column final_temperature or oven_temperature because they are to much correlated and is useless to much information

if "final_temperature" in learn_data_copy.columns:
    learn_data_copy.drop(["final_temperature"], axis=1, inplace=True)
```

```python
#Base:
#Q1 = learn_data_copy['baking_duration'].quantile(0.25)
#Q3 = learn_data_copy['baking_duration'].quantile(0.75)
#IQR = Q3 - Q1
#lower = Q1 - 1.5 * IQR
#upper = Q3 + 1.5 * IQR
#learn_data_copy = learn_data_copy[(learn_data_copy['baking_duration'] >= lower) & (learn_data_copy['baking_duration'] <= upper)]

#Egg temperature :

learn_data_copy = learn_data_copy[(learn_data_copy['egg_temperature'] <= 400)]

print(learn_data_copy['egg_temperature'].max())  # highest value

#Vanilla extract :

learn_data_copy = learn_data_copy[(learn_data_copy['vanilla_extract'] <= 14)]

print(learn_data_copy['vanilla_extract'].max())  # highest value
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math

# Seleciona apenas colunas numéricas
numeric_columns = learn_data_copy.select_dtypes(include='number').columns

# Remove colunas problemáticas (como notes_baker, se quiseres)
numeric_columns = [col for col in numeric_columns if col != 'notes_baker']

# Define o número de colunas e linhas da grelha
num_cols = 3  # 3 boxplots por linha
num_rows = math.ceil(len(numeric_columns) / num_cols)

# Cria a figura
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 4 * num_rows))

# Flatten (caso só haja 1 linha)
axes = axes.flatten()

# Cria cada boxplot
for i, col in enumerate(numeric_columns):
    sns.boxplot(x=learn_data_copy[col], ax=axes[i], color='lightblue')
    axes[i].set_title(col)

# Esconde eixos vazios (se o número não for múltiplo de 3)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Ajusta espaçamento
plt.tight_layout()
plt.show()
```
